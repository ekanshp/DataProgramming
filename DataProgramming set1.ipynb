        public string[] ArrangeString(string[] list)
        {
            int i = 1;
            string last = list[0];
            string first = list[0];
            String[] result = new String [2];
            int len = list.Length;
            while(i < len)
            {
                int res = String.Compare(last, list[i]);

                if (res < 0)
                {
                    last = list[i];
                }

                res = String.Compare(first, list[i]);
                if (res > 0)
                {
                    first = list[i];
                }

                i = i + 1;
            }
            result[0] = first;
            result[1] = last;
                
            return result;
        }

        public double triangleArea(int a,int b, int c)
        {
            double area = 0;
            float s;

            s = (a + b + c) / 2;
            area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));

### Question 6 Python
You can turn a word into pig-Latin using the following two rules (simplified):
 If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
 If the word starts with a vowel, simply append 'way' to the end of the word.
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case however.
>>> pig('happy')
'appyhay'
>>> pig('Enter')
'enterway'

```python
#Solution 6

def pig(str):
    vowels = ['a','e','i','o','u']
    if str[0].lower() in vowels:
        modifyString = str + 'way'
    else:
        modifyString = str[1:] + str[0] + 'ay'
    
    print (modifyString.lower())

pig('haPpy')
pig('Enter')
```

```python
#Solution 7

def bldcount(name: str):
    
    type_A = 0
    type_B = 0
    type_AB = 0
    type_O = 0
    type_OO = 0

    with open(name, 'r') as file:
        line = file.readline()
        
    bloodGroups = line.split()

    for bloodGroupType in bloodGroups:
        if bloodGroupType == 'A':
            type_A+=1
        elif bloodGroupType == 'B':
            type_B+=1
        elif bloodGroupType == 'AB':
            type_AB+=1
        elif bloodGroupType == 'O':
            type_O+=1
        elif bloodGroupType == 'OO':
            type_OO+=1 

    print(f'type A : {type_A}')
    print(f'type B : {type_B}')
    print(f'type AB : {type_AB}')
    print(f'type O : {type_O}')
    print(f'type OO : {type_OO}')

file_name = 'bloodtype1.txt'
bldcount(file_name)
```

```python
#Solution 8

def curconv(currency: str, amount: float) -> float:
    with open('currencies.txt', 'r') as f:
        currecies_details = f.readlines()

    for currency_detail in currecies_details:

        if currency_detail.split()[0] == currency:

            conversionRate = float(currency_detail.split()[1])
            amount_in_USD = amount * conversionRate
            break

        else:
            continue

    return amount_in_USD

print(curconv('NZD', 1234))
```
